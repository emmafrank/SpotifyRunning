more selective EDA after general investigation

choosing to only look at high tempo songs

```python
def cat_tempo(index):
    if index in running_num:
        return "running"
    elif ds.loc[index, 'tempo'] > 200:
        return "prestissimo"
    elif ds.loc[index, 'tempo'] > 168:
        return "presto"
    elif ds.loc[index, 'tempo'] > 120:
        return "allegro"
```

```python
ds['tempo_cat']  = ""

for i, rows in ds.iterrows():
    ds.loc[i, 'tempo_cat'] = cat_tempo(i) 

scaled_songs['tempo_cat']  = ds['tempo_cat'].values
```

```python
scaled_songs
```

```python
ind = scaled_songs['tempo_cat'].map(lambda x: x in ['presto', 'prestissimo', 'running'])
high_tempo_songs = scaled_songs[ind]

high_tempo_songs
```

```python
features_2 = ['danceability', 'energy', 'loudness', 'speechiness', 'valence', 'tempo_cat']

high_tempo_songs = high_tempo_songs[features_2]

high_tempo_songs
```

Machine Learning

add a running column to the overall song dataframe

```python
def if_running(index):
    if index in running_num:
        return 1
    else:
        return  0
```

```python
ds["running"] = 0

for i, rows in ds.iterrows():
    ds.loc[i, 'running'] =  if_running(i)
    
```

Final chosen dimensions to study

```python
features_3 = ['energy', 'loudness', 'tempo_cat']

scaled_songs = scaled_songs[features_3]
scaled_songs['running'] = ds['running'].values
```

```python
ind = scaled_songs['tempo_cat'].map(lambda x: x in ['allegro', 'presto', 'prestissimo', 'running'])
high_tempo_songs = scaled_songs[ind]

high_tempo_songs
```

Linear Regression Model

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
high_tempo_songs = high_tempo_songs.drop(columns = 'tempo_cat')
```

```python
high_tempo_songs
```

```python
data = high_tempo_songs
labels = high_tempo_songs['running']
indices = high_tempo_songs.index.values 
```

```python
X_train, X_test, indices_train, indices_test = train_test_split(data, indices, test_size=0.2)

y_train, y_test = labels[indices_train], labels[indices_test]
```

```python
clf.fit(X_train.values, y_train)

prediction_of_probability = clf.predict_proba(X_test)
```
